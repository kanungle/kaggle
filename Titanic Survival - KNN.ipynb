```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split

train = pd.read_csv("train.csv")
test = pd.read_csv("test.csv")

train
```

```python
drop_cols = ["Ticket","PassengerId","Cabin","Name","Embarked"]

train_df = train.drop(drop_cols,axis=1)
train_df = train_df.dropna()

test_df = test.drop(drop_cols,axis=1)
test_df = pd.concat([test_df,test["PassengerId"]],axis=1)
test_df = test_df.dropna()

train_df['Sex'] = train_df['Sex'].replace({"male":1,"female":0})
test_df['Sex'] = test_df['Sex'].replace({"male":1,"female":0})

test_df
```

```python
target = train_df["Survived"]
predictors = train_df.drop(columns="Survived",axis=1)

model = KNeighborsClassifier(n_neighbors=1)
model.fit(predictors,target)

print(model.score(predictors,target))

test_df_anon = test_df.drop("PassengerId",axis=1)
survived = model.predict(test_df_anon)
survived = pd.DataFrame(survived,columns=["Survived"])
result = pd.merge(test_df, survived, left_index=True, right_index=True)

result
```
