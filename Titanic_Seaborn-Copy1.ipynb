```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("train.csv")
df
```

```python
plt1 = sns.boxplot(data=df,x='Sex',y='Age')
plt1.set_title("Age Distribution by Gender")
```

```python
men = df[df['Sex']=='male']['Age']
women = df[df['Sex']=='female']['Age']

sns.kdeplot(men,shade=True)
sns.kdeplot(women,shade=True)
plt.legend(labels=['men','women'])
```

```python
sns.pairplot(df)
```

```python
sns.violinplot(data=df, x='Sex',y='Age')
```
